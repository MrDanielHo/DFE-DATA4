# Pillars of AWS Well-Architectured Framework
## 1. Operational Excellence
1. Operations-as-Code
    * The entire workload i.e. development of applications together with the infrastructure, should consist exclusively of code.<br>
    * This way, operational errors can be reduced, and the execution, as can updates or changes, can be automated, saving time and resources.
2. Frequent but small and reversible changes
3. Evolve procedures alongside the workload
    * As software development progresses, so should the assoicated processes.
    * Regular and dedicated routines can help to find improvement opportunities, like finding processes that can be automated, and validate the effectiveness of the procedures.
4. Failure prevention
    * Identify sources of failures and remove the cause before issues reoccur by performing preventitive exercises.
    * Regular events and simulated exercises will increase awareness inside the team and improve reactions times.
5. Learning from operationa failures
    * Every incident and failure is a lesson to analyse and learn from.
    * Lessons and appropriate improvements should be well documented and shared throughout the whole company.

## Security
1. Implementation of a strong identity foundation
    * Users should be granted only as much access as they need to fulfil a task.
    * Appropriate authorisation should be a major requirement for all resources in the cloud system.
2. Traceability
    * Real-time monitoring, alerting, and auditing actions and changes in the environment should be logged and made available for automatic investigation and action routines.
3. Apply security at all levels
    * Defensive strategies should be applied to all posibble levels.
4. Security by automation
    * By using automated security mechanisms, architectures gain the ability to scale faster and more cost-effectively
5. Data protection
    * Make the use of data encryption, tokenisation, and access control, mandatory.
6. Access
    * Reduce the risk of improper handling or modification of sensitive data by users by preventing direct access when not needed.
7. Incident preparation
    * Prepare incident mangement and investigation policies in advance to handle incidences when they occur.

## Reliability
1. Automatic recovery
    * Identify potential outages by using Key Performance Indicators (KPIs) on the workload that will trigger the monitoring system.
        * This allows the appropriate action to either prevent the outage or automatically begin remediation.
2. Test recovery procedures
    * Develop recovery strategies by testing different failure scenarios.
    * Other than an on-premise environment, the cloud allows simulating various scaled scenarios.
3. Horizontal scaling for better availability
    * Prevent common points of failure by scaling the infrastructure horizontally, i.e. distributing requests across multiple resources rather than hoarding the entire workload on a single resource.
4. Stop guessing capacity
    * Use monitoring as a tool to detect demand and scale the environment accordingly by automated addition or removal.
5. Manage changes in automation
    * Changes to the infrastructure should only be made by automated and trackable actions.

## Performance and Efficiency.
1. Make use of advande technologies
    * Using Software-as-a-Service instead of hosting, operating, and managing a tool provides more free time and resources for the development team.
2. Global in minutes
    * AWS Regions allows workloads to be deployed around the world, at various scales, reducing latency for customers.
3. Serverless architectures
    * Many traditional computing activities can be realised through serverless solutions that can be maintained without the need of a physical server, saving resources for operations and management.
4. Experiment more
    * Various tests can be performaed with virtual resources to determing which service or resource and which type of configuration is best suited for individuals' requirements.
5. Know available options - make the right choices
    * Make sure you know about the different services, and potentially the service that aligns best with individual workload goals.
        * For example, when making decisions, consider the appropriate database/storage concept that best suits the needs.

## Cost optimisation
1. Implement cloud financial management
    * Build capabilities to manage and spread awareness of costs and expenses in the cloud environment.
2. Adpot a consumption model
    * Pay-as-you-go
    * Analyse the actual business needs and match resources to current requirement.
3. Measure overall efficiency
    * Meaure workload, business performance, and the costs associated with deployment.
    * Use these metrics to determine the gains you make by increasing performance and reducing costs.
4. Stop spending money on data centre operations
    * AWS takes over traditional data centre tasks and does not charge extra to manage and update operating systems or applications with managed services.
5. Analyse and attribute expenses
    * AWS makes it easy to identify system usage and costs, and transparently allocate IT costs to individual workload owners based on this data.
    * This helps masure return on investment (ROI) and enables workload owners to optimise resources and reduce costs.
