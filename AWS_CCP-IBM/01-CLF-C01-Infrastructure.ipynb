# AWS Global Infrastructure: Availability Zones, Regions, Edge Locations, Regional Edge Caches

The componenets that make up the Amazon Web Service (AWS) Global Infrastructure are:
*   Availability Zones (AZs),
*   Regions,
*   Edge Locations, and
*   Regional Edge Caches.

## Availability Zones
Availability Zones and Regions are closely related.<br>
AZs are essentially the physical data centres of AWS.<br>
This is where the actual compute, storage, network, and database resources are hosted that are provisioned within Virtual Private Clouds (VPCs).<br>
A common misconception is that a single availability zone is equal to a single data centre, this is not the case.<br>
It is likely that multiple data centres located close together form a single availability zone.

Each AZ will always have at least one other AZ that is geographically located within the same area, usually a city, linked by highly resilent and very low latency private fiber optic connections.<br>
However, each AZ will be isolated from the others using separate power and network connectivity that minimises impact to other AZs should a single AZ fail.

These low latency links between AZs are used by many AWS services to replicate data for high availability and resilience purposes.<br>
For example when a Relational Database Service (RDS) is configured for *Multi-AZ* deployments, AWS will use synchronous replication between its primary and secondary database and asynchronous replication for any read replicas that have been created.

There are typically between three to five AZs linked together via these low latency connections.<br>
This localised geographical grouping of multiple AZs, including multiple data centres, is defined as an AWS region.

Multiple AZs within a region allows you to create hightly available and resilent applications and services.<br>
By architecting your solutions to utilise resources across a number of AZs ensures that minimal to no impact will occur to your infrastructure should an AZ experience failure.

Anyone can deploy resources in the cloud, but architecting them in a way that ensures your infrastructure remains stable, available, and resilient when faced with a disaster is a different matter.<br>
Making use of two or more AZs in a region helps you maintain high availablity of your infrastructure and its always recommended practice.

## Regions
As previously defined, a Region is a collection of availability zones that are geographically located close to one other.<br>
This is generally indicated by AZs within the same city.<br>
AWS has deploted them across the globe to allow its worldwide customer base to take advantage of low latency connections.<br>
Every Region will act independently of the others, and each will contain at least two Availbility Zones.

For example, if an organisation based in London has customers throughout Europe, it would make no logical sense to deploy services in the Sydney Region simply due to the latency response times for its customer.<br>
Instead, the company would select the region most approp4riate for them and their customer base, which may be the London, Frankfurt, or Ireland Region.

Having global regions also allows for compliance with regulations, laws, and governance relating to data storage (at rest and in transit).<br>
For example, you may be required to kepp all data within a specific location, such as Europe.<br>
Having multiple regions within this location allows an organisation to meet this requirement.

Similarly to how utilising multiple AZs within a region creates a level of high availability, the same can be applied to utilising multiple regions.<br>
Depending on the level of business continuity you require, you may choose to architect your AWS environment to support your applications and services across multiple regions, should an entire region become unavailable, perhaps due to a natural disaaster.

You may want to use multiple regions if you are a global organisation serving customers in different counrtries that have specific laws and governance about the use of data.<br>
In this case, you could even connect different VPCs together in different regions.

The number of regions is increasing year after year as AWS works to keep up with the demand for cloud computing services.<br>
There are currently 20 Regions and 61 Availability Zones, with 4 Regions and 12 AZs planned.

Not all AWS services are available in every region.<br>
This is a consideration that must be taken into account when architecting your infrastructure.<br>
Some services are classed as global services, such as AWS Identity and Access Management (IAM) or Amazon CloudFront, which means that these services are not tied to a specific region.<br>
However, most services are region-specific, and it's down to the user to understand which services are (un)availble in which region.

Every single region is independant from other regions.<br>
The AWS GovCloud is an isolated region in the US that is only available to US government agencies and organisations in government-regulated industries, which must meet strict requirements.

## Edge Locations
Edge Locations are AWS sites deployed in major cities and highly populated areas across the globe.<br>
They far outnumber the number of availability zones available.

While Edge Locations are not used to deploy main infrastructures such as EC2 instances, EBS storage, VPCs, or RDS resources like AZs, they are used by AWS services such as Amazon CloudFront and AWS Lambda@Edge to cache data and reduce latency for end-user access by using the Edge Locations as a global Content Delivery Network (CDN).

As a result, Edge Locations are primarily used by end users who are accessing and using your services.<br>
For example, you maky have your website hosed on EC2 instances and S3 (your origin) within the Ohio region with a configured CloudFront distribution associated.<br>
When a user accessess your website from Europe, they would be redirected to their closest Edge Location (in Europe) where cached data could be read on your website, significantly reducing latency.

## Regional Edge Cache
In November 2016, AWS introduced Regional Edge Cache, a new type of Edge Location.<br>
These sit between your CloudFront Origin servers and the Edge Locations.<br>
A Regional Edge Cache has a larger cache-width than each of the individual Edge Locations, and because data expires from the cache at the Edge Locations, the data is retained at the Regional Edge Caches.

Therefore when data is requrest at the Edge Location that is no longer available, the Edge Location can retrieve the cached data from the Regional Edge Cache instead of the Origin servers, which would have higher latency.
