```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_melted = pd.melt(boston_df)
```

```python
fig, ax = plt.subplots()
ax = sns.boxplot(x='variable', y='value', data=data_melted)
ax.set_xlabel("")
ax.set_ylabel('MEDV')
plt.show
```

Looking at the magnitude of the box plot:

We have two variables in the 100s (Tax & B) and the rest towards zero.<br>
This poses an issue if we used a regressor to see which features best described the model, these 2 features will affect the model performance.<br>
They will dominate the model in respect to the other features as they would be seen with higher weight.

Age and tax are highly skewed.<br>
Age has a high negative skew - where the median is greater than the mean. This could be due to Boston being an old city, the median house age is greater than the mean.

Tax has a high positive skew - a small population are paying higher taxes whereas the majority are paying lower taxes.

We will need to homogenise the data before feeding a machine learning model with it.<br>
We do not want an extreme value to dominate over another variable and therefore, this plot shows us that scaling is crucial for our purposes.<br>
We need to transform the data before feeding it into a model.<br>
We need to scale the data as they have features with different magnitudes.<br>
Scaling will remove skewness and will help assign equal weight to different features.<br>
The downside of scaling it hat it will remove some physical interpretability.

```python
'''
We need to remove MEDV from the preprocessing because this is the column we want to predict target.
'''

X = boston_df.drop(['MEDV'], axis=1)
y = boston_df['MEDV']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_scaled = scaler.transform(X)
```

```python
df_scaled = pd.DataFrame(X_scaled, columns=list(X))
```

```python
df_scaled_melted = pd.melt(df_scaled)
```

```python
fig, ax = plt.subplots()
ax = sns.boxplot(x='variable', y='value', data=df_scaled_melted)
ax.set_xlabel("")
ax.set_ylabel('MEDV')
plt.show
```

After standardising the data, the median values are zero and data is set around that.<br>
The distribution of data has remained the same.

We have performed the following steps:
1. Initialise the transformer `scaler = StandardScaler()`
2. Fit the data `scaler.fit(df)`
3. Transform the data `scaler.transform(df)`

Instead of these 3 steps, we can use `scaler.fit(df)`.

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler

# we can use a single pipeline class to define multiple steps for our pipeline.

transformer = Pipeline(
    [
        ('scaler', StandardScaler()),
        ('robust', RobustScaler())
    ]
)

transformer.fit(X)
pd.DataFrame(transformer.transform(X), columns=list(X)).head()
```
