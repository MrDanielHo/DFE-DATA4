```python
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder()
ct = make_column_transformer(
    (ohe, ['smoker', 'sex', 'day', 'time']),
    remainder='passthrough'
)
```

```python
data_new = pd.DataFrame(ct.fit_transform(df))
```

# Car Insurance

```python
# toy_dataset
data_toy = pd.DataFrame({
    'constructor': ['Toyota', 'Audi', 'Renault']
})
data_toy
```

```python
ohe_toy = OneHotEncoder()
pd.DataFrame(ohe_toy.fit_transform(data_toy).toarray())
```

```python
'''
What happens if we had another manufacturer added to the list:
'''

new_data = pd.DataFrame(
    {'constructor': ['Toyota', 'Jeep']}
)
```

```python
'''
We want to put the new manufacturing into our original list:
'''

ohe_toy.transform(new_data)
```

```python
'''
As per the above, the new manufacture is unknown to the dataframe and we cannot insert it.

But we can take the below to include the data:
'''

ohe_toy = OneHotEncoder(handle_unknown='ignore')
pd.DataFrame(ohe_toy.fit_transform(data_toy).toarray())
```

```python
ohe_toy.transform(new_data)
```

```python
ohe_toy = OneHotEncoder(handle_unknown='ignore', sparse=False)
pd.DataFrame(ohe_toy.fit_transform(data_toy))
```

```python
pd.DataFrame(ohe_toy.transform(new_data))
```

We now have an extra row for the Jeep made of zeroes.<br>
None of the the categories observed in the training is suitable to classify that example.<br>
i.e. we do not have Jeeps but it does belong to another category.
